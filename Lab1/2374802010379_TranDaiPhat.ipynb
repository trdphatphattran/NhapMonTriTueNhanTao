## Cơ Sở Lý Thuyết

### 1. Tìm Kiếm Mù
Tìm kiếm mù (Uninformed Search) là các thuật toán tìm kiếm không sử dụng thông tin heuristic, duyệt qua không gian trạng thái theo cách có hệ thống, dựa trên cấu trúc bài toán.

### 2. Tìm Kiếm theo Chiều Rộng (BFS)
- **Nguyên lý**: Khám phá tất cả các nút ở mức hiện tại trước khi chuyển sang mức tiếp theo.
- **Cấu trúc dữ liệu**: Hàng đợi (Queue).
- **Đặc điểm**:
  - Trong đồ thị không trọng số, BFS đảm bảo tìm đường đi ngắn nhất (theo số cạnh).
  - Trong đồ thị có trọng số, BFS không đảm bảo đường đi có tổng trọng số nhỏ nhất, vì nó không xem xét trọng số của các cạnh.

### 3. Tìm Kiếm theo Chiều Sâu (DFS)
- **Nguyên lý**: Khám phá một nhánh đến tận cùng trước khi quay lại thử nhánh khác.
- **Cấu trúc dữ liệu**: Ngăn xếp (Stack) hoặc đệ quy.
- **Đặc điểm**:
  - DFS không đảm bảo đường đi ngắn nhất trong cả đồ thị không trọng số và có trọng số.
  - Kết quả phụ thuộc vào thứ tự duyệt các nút kề.

### 4. Đồ Thị Không Trọng Số và Có Trọng Số
- **Đồ thị không trọng số**: Các cạnh có trọng số ngầm định bằng 1. BFS tối ưu theo số cạnh.
- **Đồ thị có trọng số**: Các cạnh có trọng số khác nhau (ví dụ: chi phí, khoảng cách). BFS và DFS không được thiết kế để tối ưu hóa tổng trọng số, dẫn đến kết quả không tối ưu.

### 5. Bài Toán 
Cho các đồ thị vô hướng (không trọng số và có trọng số) biểu diễn mạng lưới hoặc mê cung. Nhiệm vụ là tìm đường đi từ nút bắt đầu (**S**) đến nút đích (**G** hoặc **H**) bằng BFS và DFS, thông qua minh họa thủ công và lập trình với chú thích chi tiết.

**Minh họa DFS** :
1. Khởi tạo: Ngăn xếp = [(S, [S])], Đã thăm = {S}

3. Lấy A, thêm D: Ngăn xếp = [(D, [S, A, D])], Đã thăm = {S, A, D}
4. Lấy D, thêm F: Ngăn xếp = [(F, [S, A, D, F])], Đã thăm = {S, A, D, F}
5. Lấy F, không có kề mới: Ngăn xếp rỗng, quay lại S.
6. Từ S, thêm B: Ngăn xếp = [(B, [S, B])], Đã thăm = {S, A, D, F, B}
7. Lấy B, thêm E: Ngăn xếp = [(E, [S, B, E])], Đã thăm = {S, A, D, F, B, E}
8. Lấy E, thêm G: G là đích, trả về [S, B, E, G]

**Kết quả**: Đường đi DFS: S → B → E → G.

### Phần 3: Triển Khai Mã Python

#### BFS trên Đồ Thị Không Trọng Số
Chạy ô mã dưới đây để kiểm chứng BFS trên **Đồ thị mẫu 1**.

````python
from collections import deque

def bfs(graph, start, goal):
    # Khởi tạo hàng đợi với nút bắt đầu và đường đi ban đầu
    queue = deque([(start, [start])])
    # Khởi tạo tập hợp các nút đã thăm để tránh lặp
    visited = set([start])
    
    # Lặp cho đến khi hàng đợi rỗng
    while queue:
        # Lấy nút đầu tiên và đường đi tương ứng từ hàng đợi
        (node, path) = queue.popleft()
        # Kiểm tra nếu nút hiện tại là đích
        if node == goal:
            return path
        # Duyệt qua các nút kề của nút hiện tại
        for neighbor in graph[node]:
            # Nếu nút kề chưa được thăm
            if neighbor not in visited:
                # Thêm nút kề vào tập đã thăm
                visited.add(neighbor)
                # Thêm nút kề và đường đi mới vào hàng đợi
                queue.append((neighbor, path + [neighbor]))
    # Trả về None nếu không tìm thấy đường đi
    return None

# Định nghĩa Đồ thị mẫu 1 (không trọng số)
graph1 = {
    'S': ['A', 'B', 'C'],
    'A': ['S', 'D'],
    'B': ['S', 'E'],
    'C': ['S'],
    'D': ['A', 'F'],
    'E': ['B', 'G'],
    'F': ['D'],
    'G': ['E']
}

# Chạy BFS và in kết quả
path = bfs(graph1, 'S', 'G')
print("Đường đi BFS trên Đồ thị mẫu 1:", path)
````

#### DFS trên Đồ Thị Không Trọng Số
Chạy ô mã dưới đây để kiểm chứng DFS trên **Đồ thị mẫu 1**.

````python
def dfs(graph, start, goal, visited=None, path=None):
    # Khởi tạo tập đã thăm và đường đi nếu chưa có
    if visited is None:
        visited = set()
    if path is None:
        path = [start]
    
    # Thêm nút hiện tại vào tập đã thăm
    visited.add(start)
    # Kiểm tra nếu nút hiện tại là đích
    if start == goal:
        return path
    
    # Duyệt qua các nút kề của nút hiện tại
    for neighbor in graph[start]:
        # Nếu nút kề chưa được thăm
        if neighbor not in visited:
            # Gọi đệ quy DFS trên nút kề với đường đi mới
            new_path = dfs(graph, neighbor, goal, visited, path + [neighbor])
            # Nếu tìm thấy đường đi, trả về
            if new_path:
                return new_path
    # Trả về None nếu không tìm thấy đường đi
    return None

# Định nghĩa Đồ thị mẫu 1 (không trọng số)
graph1 = {
    'S': ['A', 'B', 'C'],
    'A': ['S', 'D'],
    'B': ['S', 'E'],
    'C': ['S'],
    'D': ['A', 'F'],
    'E': ['B', 'G'],
    'F': ['D'],
    'G': ['E']
}

# Chạy DFS và in kết quả
path = dfs(graph1, 'S', 'G')
print("Đường đi DFS trên Đồ thị mẫu 1:", path)
````

#### BFS trên Đồ Thị Có Trọng Số
Chạy ô mã dưới đây để kiểm chứng BFS trên **Đồ thị mẫu 5**.

````python
from collections import deque

def bfs_weighted(graph, start, goal):
    # Khởi tạo hàng đợi với nút bắt đầu, đường đi, và tổng trọng số
    queue = deque([(start, [start], 0)])
    # Khởi tạo tập hợp các nút đã thăm
    visited = set([start])
    
    # Lặp cho đến khi hàng đợi rỗng
    while queue:
        # Lấy nút, đường đi, và tổng trọng số từ đầu hàng đợi
        (node, path, total_weight) = queue.popleft()
        # Kiểm tra nếu nút hiện tại là đích
        if node == goal:
            return path, total_weight
        # Duyệt qua các nút kề và trọng số tương ứng
        for neighbor, weight in graph[node]:
            # Nếu nút kề chưa được thăm
            if neighbor not in visited:
                # Thêm nút kề vào tập đã thăm
                visited.add(neighbor)
                # Thêm nút kề, đường đi mới, và tổng trọng số cập nhật vào hàng đợi
                queue.append((neighbor, path + [neighbor], total_weight + weight))
    # Trả về None và 0 nếu không tìm thấy đường đi
    return None, 0

# Định nghĩa Đồ thị mẫu 5 (có trọng số)
graph5 = {
    'S': [('A', 1), ('B', 2), ('C', 3)],
    'A': [('S', 1), ('D', 4), ('G', 3)],
    'B': [('S', 2), ('G', 5)],
    'C': [('S', 3)],
    'D': [('A', 4), ('G', 6)],
    'G': [('B', 5), ('D', 6), ('A', 3)]
}

# Chạy BFS và in kết quả
path, weight = bfs_weighted(graph5, 'S', 'G')
print("Đường đi BFS trên Đồ thị mẫu 5:", path, "Tổng trọng số:", weight)
````

#### DFS trên Đồ Thị Có Trọng Số
Chạy ô mã dưới đây để kiểm chứng DFS trên **Đồ thị mẫu 5**.

````python
def dfs_weighted(graph, start, goal, visited=None, path=None, total_weight=0):
    # Khởi tạo tập đã thăm và đường đi nếu chưa có
    if visited is None:
        visited = set()
    if path is None:
        path = [start]
    
    # Thêm nút hiện tại vào tập đã thăm
    visited.add(start)
    # Kiểm tra nếu nút hiện tại là đích
    if start == goal:
        return path, total_weight
    
    # Duyệt qua các nút kề và trọng số tương ứng
    for neighbor, weight in graph[start]:
        # Nếu nút kề chưa được thăm
        if neighbor not in visited:
            # Gọi đệ quy DFS trên nút kề với đường đi và tổng trọng số mới
            new_path, new_weight = dfs_weighted(graph, neighbor, goal, visited, path + [neighbor], total_weight + weight)
            # Nếu tìm thấy đường đi, trả về
            if new_path:
                return new_path, new_weight
    # Trả về None và 0 nếu không tìm thấy đường đi
    return None, 0

# Định nghĩa Đồ thị mẫu 5 (có trọng số)
graph5 = {
    'S': [('A', 1), ('B', 2), ('C', 3)],
    'A': [('S', 1), ('D', 4), ('G', 3)],
    'B': [('S', 2), ('G', 5)],
    'C': [('S', 3)],
    'D': [('A', 4), ('G', 6)],
    'G': [('B', 5), ('D', 6), ('A', 3)]
}

# Chạy DFS và in kết quả
path, weight = dfs_weighted(graph5, 'S', 'G')
print("Đường đi DFS trên Đồ thị mẫu 5:", path, "Tổng trọng số:", weight)
````

# Trả lời câu hỏi lý thuyết ở đây:

# 1. 
BFS đảm bảo đường đi ngắn nhất trên đồ thị không có trọng số nhưng không đảm bảo trên đồ thị có trọng số vì:
- Không trọng số: mọi cạnh đều có trọng số bằng nhau (có thể là 1 đơn vị), BFS sẽ xét các đỉnh cách đỉnh bắt đầu 1 bước, sau đó 2 bước, ...(đi những đỉnh gần trước, đỉnh xa sau). Vì thế nên BFS đến đỉnh nào trước tiên thì đó chính là đường đi ngắn nhất.
- Có trọng số: các cạnh có thể có trọng số khác nhau (có cạnh ngắn, cạnh dài) mà BFS không cần đến trọng số cạnh, nó chỉ xét số lượng bước đi, vì thế nên BFS có thể đi con đường ít bước nhưng trọng số lớn hơn, thay vì đi con đường nhiều bước nhưng trọng số ít hơn.

# 2. 
Để BFS và DFS tránh lặp vô hạn thì cần:
- DFS: đi sâu theo nhánh, nhưng nếu gặp lại đỉnh đã thăm thì DFS sẽ quay lại ngay, không đi tiếp nữa.
- BFS: đi theo hàng, cần kiểm tra xem đỉnh sắp vào hàng đợi đã thăm hay chưa, nếu rồi thì không thêm lại vào hàng đợi.

Chạy BFS trên **Đồ thị mẫu 2** với chú thích chi tiết.

````python
# code here
from collections import deque

def bfs(graph, start, goal):
    # Khởi tạo hàng đợi với nút bắt đầu và đường đi ban đầu
    queue = deque([(start, [start])])
    # Khởi tạo tập hợp các nút đã thăm để tránh lặp
    visited = set([start])
    
    # Lặp cho đến khi hàng đợi rỗng
    while queue:
        # Lấy nút đầu tiên và đường đi tương ứng từ hàng đợi
        (node, path) = queue.popleft()
        # Kiểm tra nếu nút hiện tại là đích
        if node == goal:
            return path
        # Duyệt qua các nút kề của nút hiện tại
        for neighbor in graph[node]:
            # Nếu nút kề chưa được thăm
            if neighbor not in visited:
                # Thêm nút kề vào tập đã thăm
                visited.add(neighbor)
                # Thêm nút kề và đường đi mới vào hàng đợi
                queue.append((neighbor, path + [neighbor]))
    # Trả về None nếu không tìm thấy đường đi
    return None

# Định nghĩa Đồ thị mẫu 2 (không trọng số)
graph2 = {
    'S': ['A', 'B'],
    'A': ['B', 'C'],
    'B': [ 'D'],
    'C': ['D'],
    'D': ['G'],
    'G': []
}

# Chạy BFS và in kết quả
path = bfs(graph2, 'S', 'G')
print("Đường đi BFS trên Đồ thị mẫu 2:", path)
````

### Sửa mã BFS để đếm số nút đã thăm trên **Đồ thị mẫu 4**. In ra kết quả và đảm bảo mã có chú thích đầy đủ.

````python
# code here
from collections import deque

def bfs(graph, start, goal):
    # Khởi tạo hàng đợi với nút bắt đầu và đường đi ban đầu
    queue = deque([(start, [start])])
    # Khởi tạo tập hợp các nút đã thăm để tránh lặp
    visited = set([start])
    # Đếm số nút đã thăm
    count = 1
    # Lặp cho đến khi hàng đợi rỗng
    while queue:
        (node, path) = queue.popleft()
        # Kiểm tra nếu nút hiện tại là đích
        if node == goal:
            print("Số nút đã thăm:", count)
            return path
        # Duyệt qua các nút kề của nút hiện tại
        for neighbor in graph[node]:
            # Nếu nút kề chưa được thăm
            if neighbor not in visited:
                # Thêm nút kề vào tập đã thăm
                visited.add(neighbor)
                # Thêm nút kề và đường đi mới vào hàng đợi
                queue.append((neighbor, path + [neighbor]))
                count += 1  # Tăng biến đếm khi gặp nút chưa thăm
    # Trả về None nếu không tìm thấy đường đi
    print("Số nút đã thăm:", count)
    return None

# Định nghĩa Đồ thị mẫu 4 (không trọng số)
graph4 = {
    'S': ['A', 'C'],
    'A': ['B', 'D'],
    'B': [ 'E'],
    'C': ['F'],
    'D': ['E'],
    'E': ['G'],
    'F': ['G'],
    'G': []
}

# Chạy BFS và in kết quả
path = bfs(graph4, 'S', 'G')
print("Đường đi BFS trên Đồ thị mẫu 4:", path)
````

````python
# code here

from collections import deque
import networkx as nx
import matplotlib.pyplot as plt

def bfs(graph, start, goal):
    queue = deque([(start, [start])])
    visited = set([start])
    count = 1

    while queue:
        (node, path) = queue.popleft()
        if node == goal:
            print("Số nút đã thăm:", count)
            return path
        for neighbor in graph[node]:
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append((neighbor, path + [neighbor]))
                count += 1
    print("Số nút đã thăm:", count)
    return None

# Định nghĩa đồ thị mẫu 4 (có hướng)
graph4 = {
    'S': ['A', 'C'],
    'A': ['B', 'D'],
    'B': ['E'],
    'C': ['F'],
    'D': ['E'],
    'E': ['G'],
    'F': ['G'],
    'G': []
}

path = bfs(graph4, 'S', 'G')
print("Đường đi BFS trên Đồ thị mẫu 4:", path)

# Tạo đồ thị có hướng
G = nx.DiGraph()

# Thêm các cạnh
for node in graph4:
    for neighbor in graph4[node]:
        G.add_edge(node, neighbor)

# Layout cố định 
pos = nx.spring_layout(G, seed=42)

# Vẽ toàn bộ đồ thị
nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color='gray',
        node_size=1000, font_size=12, arrowsize=20)

# Nếu tìm thấy đường đi, vẽ nổi bật
if path:
    # Các cạnh thuộc đường đi
    path_edges = list(zip(path, path[1:]))
    # Vẽ các cạnh của đường đi màu đỏ
    nx.draw_networkx_edges(G, pos, edgelist=path_edges, edge_color='red', width=3, arrows=True, arrowsize=25)
    # Vẽ các nút trong đường đi màu cam
    nx.draw_networkx_nodes(G, pos, nodelist=path, node_color='orange', node_size=1000)

# Tiêu đề và hiển thị
plt.title("Đường đi BFS")
plt.show()
````

2. Lấy S, thêm A: Ngăn xếp = [(A, [S, A])], Đã thăm = {S, A}
3. Lấy A, thêm B: Ngăn xếp = [(B, [S, A, B])], Đã thăm = {S, A, B}
4. Lấy B, thêm C: Ngăn xếp = [(C, [S, A, B, C])], Đã thăm = {S, A, B, C}
5. Lấy C, thêm F: Ngăn xếp = [(F, [S, A, B, C, F])], Đã thăm = {S, A, B, C, F}
6. Lấy F, thêm G: Ngăn xếp = [(G, [S, A, B, C, F, G])], Đã thăm = {S, A, B, C, F, G}
7. Lấy G, thêm H: Ngăn xếp = [(H, [S, A, B, C, F, G, H])], Đã thăm = {S, A, B, C, F, G, H}
8. Lấy H: H là đích, trả về [S, A, B, C, F, G, H]
Kết luận: Đường đi DFS: S -> A -> B -> C -> F -> G -> H

````python
# code here
from collections import deque

def bfs_weighted(graph, start, goal):
    # Khởi tạo hàng đợi với nút bắt đầu, đường đi, và tổng trọng số
    queue = deque([(start, [start], 0)])
    # Khởi tạo tập hợp các nút đã thăm
    visited = set([start])
    
    # Lặp cho đến khi hàng đợi rỗng
    while queue:
        # Lấy nút, đường đi, và tổng trọng số từ đầu hàng đợi
        (node, path, total_weight) = queue.popleft()
        # Kiểm tra nếu nút hiện tại là đích
        if node == goal:
            return path, total_weight
        # Duyệt qua các nút kề và trọng số tương ứng
        for neighbor, weight in graph[node]:
            # Nếu nút kề chưa được thăm
            if neighbor not in visited:
                # Thêm nút kề vào tập đã thăm
                visited.add(neighbor)
                # Thêm nút kề, đường đi mới, và tổng trọng số cập nhật vào hàng đợi
                queue.append((neighbor, path + [neighbor], total_weight + weight))
    # Trả về None và 0 nếu không tìm thấy đường đi
    return None, 0

# Định nghĩa Đồ thị mẫu 6 (có trọng số)
graph6 = {
    'S': [('A', 2), ('C', 5)],
    'A': [('S', 2), ('B', 3), ('D', 4)],
    'C': [('S', 5), ('D', 7), ('F', 9)],
    'B': [('A', 3), ('E', 6)],
    'D': [('A', 4), ('C', 7), ('E', 8)],
    'E': [('H', 10), ('D', 8), ('B', 6)],
    'F': [('G', 12), ('C', 9)],
    'G': [('H', 15), ('F', 12)],
    'H': [('G', 15), ('E', 10)]
}

# Chạy BFS và in kết quả
path, weight = bfs_weighted(graph6, 'S', 'H')
print("Đường đi BFS trên Đồ thị mẫu 6:", path, "Tổng trọng số:", weight)

# code here
def dfs_weighted(graph, start, goal, visited=None, path=None, total_weight=0):
    # Khởi tạo tập đã thăm và đường đi nếu chưa có
    if visited is None:
        visited = set()
    if path is None:
        path = [start]
    
    # Thêm nút hiện tại vào tập đã thăm
    visited.add(start)
    # Kiểm tra nếu nút hiện tại là đích
    if start == goal:
        return path, total_weight
    
    # Duyệt qua các nút kề và trọng số tương ứng
    for neighbor, weight in graph[start]:
        if neighbor not in visited:
            new_path, new_weight = dfs_weighted(graph, neighbor, goal, visited, path + [neighbor], total_weight + weight)
            if new_path:
                return new_path, new_weight
    # Trả về None và 0 nếu không tìm thấy đường đi
    return None, 0

# Định nghĩa Đồ thị mẫu 6 (có trọng số)
graph6 = {
    'S': [('A', 2), ('C', 5)],
    'A': [('S', 2), ('B', 3), ('D', 4)],
    'C': [('S', 5), ('D', 7), ('F', 9)],
    'B': [('A', 3), ('E', 6)],
    'D': [('A', 4), ('C', 7), ('E', 8)],
    'E': [('H', 10), ('D', 8), ('B', 6)],
    'F': [('G', 12), ('C', 9)],
    'G': [('H', 15), ('F', 12)],
    'H': [('G', 15), ('E', 10)]
}

# Chạy DFS và in kết quả
path, weight = dfs_weighted(graph6, 'S', 'H')
print("Đường đi DFS trên Đồ thị mẫu 6:", path, "Tổng trọng số:", weight)
````

### Đồ thị BFS và DFS trên đồ thị mẫu 7

````python
# code here
from collections import deque

def bfs(graph, start, goal):
    # Khởi tạo hàng đợi với nút bắt đầu và đường đi ban đầu
    queue = deque([(start, [start])])
    # Khởi tạo tập hợp các nút đã thăm để tránh lặp
    visited = set([start])
    
    # Lặp cho đến khi hàng đợi rỗng
    while queue:
        # Lấy nút đầu tiên và đường đi tương ứng từ hàng đợi
        (node, path) = queue.popleft()
        # Kiểm tra nếu nút hiện tại là đích
        if node == goal:
            return path
        # Duyệt qua các nút kề của nút hiện tại
        for neighbor in graph[node]:
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append((neighbor, path + [neighbor]))
    # Trả về None nếu không tìm thấy đường đi
    return None

graph7 = {
    'S': ['A', 'D', 'E'],
    'A': ['S', 'D', 'B', 'E'],
    'B': ['A', 'E', 'C', 'F'],
    'C': ['B', 'F', 'G'],
    'D': ['A', 'S', 'E'],
    'E': ['S', 'A', 'B', 'D', 'F', 'H'],
    'F': ['B', 'C', 'E', 'G', 'H'],
    'G': ['H', 'F', 'C'],
    'H': ['E', 'F', 'G']
}

# Chạy BFS và in kết quả
path = bfs(graph7, 'S', 'H')
print("Đường đi BFS trên Đồ thị mẫu 7:", path)

# code here
def dfs(graph, start, goal, visited=None, path=None):
    # Khởi tạo tập đã thăm và đường đi nếu chưa có
    if visited is None:
        visited = set()
    if path is None:
        path = [start]
    
    # Thêm nút hiện tại vào tập đã thăm
    visited.add(start)
    # Kiểm tra nếu nút hiện tại là đích
    if start == goal:
        return path
    
    # Duyệt qua các nút kề của nút hiện tại
    for neighbor in graph[start]:
        # Nếu nút kề chưa được thăm
        if neighbor not in visited:
            # Gọi đệ quy DFS trên nút kề với đường đi mới
            new_path = dfs(graph, neighbor, goal, visited, path + [neighbor])
            # Nếu tìm thấy đường đi, trả về
            if new_path:
                return new_path
    # Trả về None nếu không tìm thấy đường đi
    return None

graph7 = {
    'S': ['A', 'D', 'E'],
    'A': ['S', 'D', 'B', 'E'],
    'B': ['A', 'E', 'C', 'F'],
    'C': ['B', 'F', 'G'],
    'D': ['A', 'S', 'E'],
    'E': ['S', 'A', 'B', 'D', 'F', 'H'],
    'F': ['B', 'C', 'E', 'G', 'H'],
    'G': ['H', 'F', 'C'],
    'H': ['E', 'F', 'G']
}

# Chạy DFS và in kết quả
path = dfs(graph7, 'S', 'H')
print("Đường đi DFS trên Đồ thị mẫu 7:", path)
````

````python
# code here
from collections import deque

def bfs_all_paths(graph, start, goal):
    # Khởi tạo hàng đợi với nút bắt đầu và đường đi ban đầu
    # Mỗi phần tử trong hàng đợi là một tuple (nút_hiện_tại, đường_đi_đến_nút_hiện_tại)
    queue = deque([(start, [start])])
    all_paths = []
    
    # Lặp cho đến khi hàng đợi rỗng
    while queue:
        # Lấy nút đầu tiên và đường đi tương ứng từ hàng đợi
        (node, path) = queue.popleft()
        
        # Kiểm tra nếu nút hiện tại là đích
        if node == goal:
            # Nếu là đích, thêm đường đi hiện tại vào danh sách all_paths
            all_paths.append(path)
            # Không return ở đây vì chúng ta muốn tìm tất cả các đường đi, nên sẽ tiếp tục duyệt các phần tử còn lại trong hàng đợi.
            
        # Duyệt qua các nút kề của nút hiện tại
        for neighbor in graph[node]:
            # Điều kiện để tránh tạo các chu trình VÔ HẠN trong một đường đi CỤ THỂ:
            # Chỉ thêm nút kề vào hàng đợi nếu nó CHƯA có trong đường đi hiện tại.
            # Điều này khác với việc sử dụng một tập 'visited' toàn cục, 
            # nơi một nút có thể đã được thăm trên một đường đi khác.
            if neighbor not in path:
                # Thêm nút kề và đường đi mới (bằng cách nối thêm neighbor vào path hiện tại) vào hàng đợi
                queue.append((neighbor, path + [neighbor]))
                
    # Trả về danh sách chứa tất cả các đường đi tìm được
    return all_paths

# Định nghĩa Đồ thị mẫu 7 (không trọng số)1``
graph7 = {
    'S': ['A', 'D', 'E'],
    'A': ['S', 'D', 'B', 'E'],
    'B': ['A', 'E', 'C', 'F'],
    'C': ['B', 'F', 'G'],
    'D': ['A', 'S', 'E'],
    'E': ['S', 'A', 'B', 'D', 'F', 'H'],
    'F': ['B', 'C', 'E', 'G', 'H'],
    'G': ['H', 'F', 'C'],
    'H': ['E', 'F', 'G']
}

# Chạy BFS để tìm tất cả các đường đi và in kết quả
all_paths_s_to_h = bfs_all_paths(graph7, 'S', 'H')
print("Tất cả các đường đi từ S đến H trên Đồ thị mẫu 7:")
if all_paths_s_to_h:
    for i, path in enumerate(all_paths_s_to_h):
        print(f"Đường đi {i+1}: {' -> '.join(path)}")
````
