## Code theo cách khác

```python
from tkinter import *
from tkinter import messagebox

root = Tk()
root.title("Game XO 4x4")

clicked = True
count = 0
buttons = []

def disableButtons():
    for row in buttons:
        for btn in row:
            btn.config(state=DISABLED)

def checkLines(player):
    for i in range(4):
        if all(buttons[i][j]["text"] == player for j in range(4)):
            for j in range(4):
                buttons[i][j].config(bg="#80ffaa")
            return True
        if all(buttons[j][i]["text"] == player for j in range(4)):
            for j in range(4):
                buttons[j][i].config(bg="#80ffaa")
            return True
    return False

def checkDiagonals(player):
    # chéo chính
    if all(buttons[i][i]["text"] == player for i in range(4)):
        for i in range(4):
            buttons[i][i].config(bg="#80ffaa")
        return True
    # chéo phụ
    if all(buttons[i][3-i]["text"] == player for i in range(4)):
        for i in range(4):
            buttons[i][3-i].config(bg="#80ffaa")
        return True
    return False

def checkWinner():
    for player in ['X', 'O']:
        if checkLines(player) or checkDiagonals(player):
            messagebox.showinfo("OX Game", f"Player {'1' if player == 'X' else '2'} WINNER!!")
            disableButtons()
            start()
            return

def checkDraw():
    if count == 16:
        messagebox.showinfo("OX Game", "DRAW!!")
        start()

def buttonClicked(button):
    global clicked, count
    if button["text"] == " ":
        button["text"] = 'X' if clicked else 'O'
        clicked = not clicked
        count += 1
        checkWinner()
        checkDraw()
    else:
        messagebox.showerror("OX Game", "Lỗi!! Vui lòng chọn lại")

def start():
    global clicked, count, buttons
    clicked = True
    count = 0
    buttons = []

    for i in range(4):
        row = []
        for j in range(4):
            btn = Button(root, text=" ", font=("Helvetica", 20), height=3, width=7,
                         bg="SystemButtonFace", command=lambda b=i, c=j: buttonClicked(buttons[b][c]))
            btn.grid(row=i, column=j)
            row.append(btn)
        buttons.append(row)

# Menu
gameMenu = Menu(root)
root.config(menu=gameMenu)
optionMenu = Menu(gameMenu, tearoff=False)
gameMenu.add_cascade(label="Options", menu=optionMenu)
optionMenu.add_command(label="Restart Game", command=start)

start()
root.mainloop()
```

## Code theo cách khác

```python
from tkinter import *
from tkinter import messagebox
import numpy as np

root = Tk()
root.title("Game XO 5x5")

BOARD_SIZE = 5
WIN_CONDITION = 5  

clicked = True
count = 0

board_state = np.zeros((BOARD_SIZE, BOARD_SIZE), dtype=int)
buttons = [[None for _ in range(BOARD_SIZE)] for _ in range(BOARD_SIZE)]

def disableButtons():
    for row in buttons:
        for btn in row:
            btn.config(state=DISABLED)

def start():
    global clicked, count, board_state
    clicked = True
    count = 0
    board_state = np.zeros((BOARD_SIZE, BOARD_SIZE), dtype=int)
    for r in range(BOARD_SIZE):
        for c in range(BOARD_SIZE):
            buttons[r][c].config(text=" ", bg="SystemButtonFace", state=NORMAL)

def check_winner_np():
    directions = [
        (1, 0),   
        (0, 1),   
        (1, 1),   
        (1, -1)   
    ]
    for r in range(BOARD_SIZE):
        for c in range(BOARD_SIZE):
            if board_state[r][c] == 0:
                continue
            for dr, dc in directions:
                try:
                    line = [board_state[r + i * dr, c + i * dc] for i in range(WIN_CONDITION)]
                    if abs(sum(line)) == WIN_CONDITION:
                        positions = [(r + i * dr, c + i * dc) for i in range(WIN_CONDITION)]
                        return np.sign(sum(line)), positions
                except IndexError:
                    continue
    return 0, []

def b_click(r, c):
    global clicked, count
    if buttons[r][c]["text"] == " ":
        if clicked:
            buttons[r][c]["text"] = "X"
            board_state[r][c] = 1
        else:
            buttons[r][c]["text"] = "O"
            board_state[r][c] = -1
        count += 1
        clicked = not clicked

        winner, positions = check_winner_np()
        if winner != 0:
            for x, y in positions:
                buttons[x][y].config(bg="#80ffaa")
            messagebox.showinfo("OX Game", f"Player {'1 (X)' if winner == 1 else '2 (O)'} WINS!")
            disableButtons()
            start()
        elif count == BOARD_SIZE ** 2:
            messagebox.showinfo("OX Game", "It's a tie!")
            start()

for r in range(BOARD_SIZE):
    for c in range(BOARD_SIZE):
        buttons[r][c] = Button(root, text=" ", font=("Helvetica", 16), height=2, width=4,
                               command=lambda r=r, c=c: b_click(r, c))
        buttons[r][c].grid(row=r, column=c)

start()
root.mainloop()
```
